### 2.1 Background

The broad goal of the coursework is to use the techniques we have learned in the course to study the solutions to a particular non linear partial differential equation called [the Sine-Gordon equation](https://en.wikipedia.org/wiki/Sine-Gordon_equation). 

Once complete, the notebook should provide a self contained report on your numerical investigations and provide relevant commentary in markdown cells around it that would enable another student to understand, run and make basic modifications to the notebook.

The Sine Gordon equation is as follows:

$ \frac{\partial\varphi^2}{\partial t^2} -\frac{\partial\varphi^2}{\partial x^2}+\sin \varphi =0, \quad (1)$

and the breather solutions we will study have the form:

$ \varphi (x,t)=4\arctan \left({\frac {{\sqrt {1-\omega ^{2}}}\;\cos(\omega t)}{\omega \;\cosh({\sqrt {1-\omega ^{2}}}\;x)}}\right), \quad (2)$

with $\omega$ a positive constant with a value less than 1. In the coursework you should set $\omega=0.5$ unless otherwise specified.

Eq. (2) (the breather solution) can be used to construct initial data for the initial value problem for Eq. (1) (the PDE). If you correctly implement the time evolution of this data, you should find that it is stable - the evolution is described exactly by the breather solution, which oscillates in time but with the envelope of the function remaining stable. This provides a very good test that the code is working correctly, meaning that you can then confidently use it for different initial data for the same PDE, for which no exact solution exists. Implementing known solutions is a very good way to test a numerical code.

The following analytic results for the breather solution in Eq. (2) may be useful:

Let $A = \sqrt {1-\omega ^{2}}$, then

$ \frac{\partial\varphi (x,t)}{\partial x}= \frac{-8\omega A^2 \cos(\omega t)\sinh(Ax)}
{2 (A \cos(\omega t))^2 +  \omega^2 +  \omega^2 \cosh{2Ax}} \quad (3)$

$ \frac{\partial\varphi (x,t)}{\partial t}= \frac{-4 A \sin(\omega t)}
{\cosh{Ax} \left(1 + \left(\frac{A \cos{wt}}{\omega \cosh{Ax}}\right)^2 \right)} \quad (4)$

## **Breather Solution Class**

Below I have created a 'Breather Solutions' Class which implements the breather solution given above (Eq 2,3 and 4.). This class is used to find values of the breather function under different conditions. When the class is instantiated it will hold the x points given in each question, the time which is set to 0 and omega which is set to 0.5 as instructed. Also, to help with ordering, the class has a variable referring to whether a solution has been found. This is set to false which will be change once solved so that the solution can then be plotted. I will now explain the methods used in this class:

- **get_dphidt_solution**

This function will implement Eq (4) above:

$$ \frac{\partial\varphi (x,t)}{\partial t}= \frac{-4 A \sin(\omega t)}
{\cosh{Ax} \left(1 + \left(\frac{A \cos{wt}}{\omega \cosh{Ax}}\right)^2 \right)}, \quad$$

This is the time derivative of the breather solution, which will return $ \frac{\partial\varphi (x,t)}{\partial t}$ at some x and t.

- **get_dphidx_solution**

This function will implement Eq (3) above:

$$ \frac{\partial\varphi (x,t)}{\partial x}= \frac{-8\omega A^2 \cos(\omega t)\sinh(Ax)}
{2 (A \cos(\omega t))^2 +  \omega^2 +  \omega^2 \cosh{2Ax}}, \quad$$

This is the x position derivative of the breather solution, which will return $\frac{\partial\varphi (x,t)}{\partial x}$ at some x and t.

- **get_phi_solution**

This function will implement Eq (2) above:

$$ \varphi (x,t)=4\arctan \left({\frac {{\sqrt {1-\omega ^{2}}}\;\cos(\omega t)}{\omega \;\cosh({\sqrt {1-\omega ^{2}}}\;x)}}\right), \quad$$

This is the $\varphi$ position of the breather solution at some given x and t. Here I have used an if statement. If the given x is just one point then it will return $\varphi$ at this point. However, if the given x is multiple points then I have used a for loop to find $\varphi$ at each x point and return an array of points.

- **plot_phi_solution**

This function ensures a solution to equation 2 has been found and then plots the x points against the $\varphi$ points found.

- **plot_dphidx_solution**

This function ensures a solution to equation 3 has been found and then plots the x points against the $\frac{\partial\varphi (x,t)}{\partial x}$ points found.

- **plot_dphidt_solution**

This function ensures a solution to equation 4 has been found and then plots the x points against the $\frac{\partial\varphi (x,t)}{\partial t}$ points found.


```python
class Breather_Solution :

    """
    Definition: a class to implement the breather solutions given above

    Attributes: initial conditions (x0, t and omega) and private attributes for the solutions

    Methods: Returns the values of the Breather function at a given time, and for a given 
    set of spatial coordinates, and its time and spatial derivative. Also plots for the solutions.
    """

    # constructor function
    def __init__(self, x0, t = 0, omega = 0.5):
        self.x = x0
        self.t = t
        self.omega = omega            
        self._solution_phi = []
        self._solution_dphidx = []
        self._solution_dphidt = []
        self._solution_solved = False 
        assert self.omega < 1 , 'omega must be a positive constant with a value less than 1'
        assert self.omega > 0 , 'omega must be a positive constant with a value less than 1'

    # find solution to time derivative (equation 4)
    def get_dphidt_solution(self):
        # define A to make calculation easier
        A = np.sqrt(1-self.omega**2)
        self._solution_dphidt = (-4 * A * np.sin(self.omega * self.t) / 
                                 (np.cosh(A * self.x) * (1 + (A * np.cos(self.omega * self.t) / (self.omega * np.cosh(A * self.x)))**2)))
        self._solution_solved = True 
        return self._solution_dphidt

    # find solution to x spatial derivative (equation 3)
    def get_dphidx_solution(self):
        A = np.sqrt(1-self.omega**2)
        self._solution_dphidx = (-8 * self.omega * A**2 * np.cos(self.omega * self.t) * np.sinh(A * self.x)) / (2 * (A * np.cos(self.omega * self.t))**2 + self.omega**2 + self.omega**2 * np.cosh(2 * A * self.x))
        self._solution_solved = True
        return self._solution_dphidx

    # find solution to breather function (equation 2)
    def get_phi_solution(self):
        # check to see whether multiple x points have been inputted
        if np.size(self.x) == 1:
            self._solution_phi = 4 * atan(sqrt(1 - self.omega**2) * cos(self.omega * self.t) / (self.omega * cosh(sqrt(1 - self.omega**2) * self.x)))
        else:
            for x in self.x:
                phi = 4 * atan(sqrt(1 - self.omega**2) * cos(self.omega * self.t) / (self.omega * cosh(sqrt(1 - self.omega**2) * x)))
                self._solution_phi.append(phi)
        self._solution_solved = True
        return self._solution_phi
    
    # plot solution to breather function (equation 2)
    def plot_phi_solution(self,line,label):
        assert self._solution_solved == True , 'No Breather Solution has been found!'
        plt.plot(self.x,self._solution_phi,line,label = label)
        plt.xlabel("x")
        plt.ylabel("f(x)");
    
    # plot solution to x spatial derivative (equation 3)
    def plot_dphidx_solution(self,line,label):
        assert self._solution_solved == True , 'No Spatial derivative Solution has been found!'
        plt.plot(self.x,self._solution_dphidx,line,label = label)
        plt.xlabel("x")
        plt.ylabel("dphidx(x)");
    
    # plot solution to time derivative (equation 4)
    def plot_dphidt_solution(self,line,label):
        assert self._solution_solved == True , 'No time derivative Solution has been found!'
        plt.plot(self.x,self._solution_dphidt,line,label = label)
        plt.xlabel("x")
        plt.ylabel("dphidt(x)");
```

- **Lagrange Polynomial - 5 points**

As degree n Lagrange polynomials agree exactly with a function f(x) at n+1 distinct points, to fit 5 points, the maximum possible order of a lagrange polynomial that would be a 4th degree polynomial. Below I have explicitly implemented the formulas:

$$L_k(x) = \prod_{i=0, i\neq k}^{n} \frac{(x - x_i)}{(x_k - x_i)}
$$

$$P_n(x) = \sum_{k=0}^{n} L_k(x) f(x_k)$$

As shown the two functions numerically looked fairly similar with sympy almost getting the same coefficients for $x^4$ and $x^2$ but very different results for $x^3$ and $x$. When plotting the scipy lagrange polynomial against the true function we can see that it fits the spike in the middle of the true function much better than the flat end points.

```python
# Still using the same x and phi points as the question before
[x1,x2,x3,x4,x5] = x_points
[f1,f2,f3,f4,f5] = phi_points
x = symbols('x')

# Implement formula for Lagrange polynomial (L)
L1 = (x - x2 ) / (x1 - x2) * (x - x3 ) / (x1 - x3)* (x - x4) / (x1 - x4) * (x - x5 ) / (x1 - x5)
L2 = (x - x1 ) / (x2 - x1) * (x - x3 ) / (x2 - x3)* (x - x4 ) / (x2 - x4)* (x - x5 ) / (x2 - x5)
L3 = (x - x1 ) / (x3 - x1) * (x - x2 ) / (x3 - x2)* (x - x4 ) / (x3 - x4)* (x - x5 ) / (x3 - x5)
L4 = (x - x1 ) / (x4 - x1) * (x - x2 ) / (x4 - x2)* (x - x3 ) / (x4 - x3)* (x - x5 ) / (x4 - x5)
L5 = (x - x1 ) / (x5 - x1) * (x - x2 ) / (x5 - x2)* (x - x3 ) / (x5 - x3)* (x - x4 ) / (x5 - x4)

# Implement formula for P
P = f1 * L1 + f2 * L2 + f3 * L3 + f4 * L4 + f5 * L5
print("Using sympy we get the polynomial", simplify(P))

# Using the scipy lagrange function
lagrange_polynomial = lagrange(x_points, phi_points)
print("Using the scipy function we get \n",lagrange_polynomial)

# Using the breather solution to find the true function
x_true = np.linspace(-10,10,100)
true_breather = Breather_Solution(x_true)
true_phi_points = true_breather.get_phi_solution()

# Plotting true function and lagrange polynomial on the same plot
true_breather.plot_phi_solution('-','true function')
plt.plot(x_true, lagrange_polynomial(x_true), '--', label="Lagrange polynomial")
breather.plot_phi_solution('o', 'points given')
plt.title('Testing Lagrange Polynomial with 5 Evenly Spaced Points')
plt.legend()
plt.grid();
```

- **Gauss Lobato - 12 nodes**

Gauss Lobato nodes concentrate the points at the ends of the interval given, as we can see below. This leads to the lagrange polynomial being much more accurate overall, especially at the exterior of the true function. Note here I have created a `gauss_lobato_nodes` function and used the fomulas below to find the x points:

$$u_i = \cos\left(\frac{\pi i}{N}\right)$$

but for a general interval $[a,b]$ their values need to be
mapped to

$$x_i = \frac{1}{2}(b - a)u_i + \frac{1}{2}(a + b)
$$


```python
# Setting initial conditions
N = 11
a = -15
b = 15

# Function to find the Gauss Lobato nodes
def gauss_lobato_nodes(N,a,b):
    indices = np.arange(N+1)
    # Implementing the forumla for u above
    u_points = np.cos(np.pi * indices / N)
    # Implementing the forumla for x above
    x_points = (b - a)/2.0 * u_points + (a + b)/2.0
    return x_points
x_points = gauss_lobato_nodes(N,a,b)
breather = Breather_Solution(x_points)
phi_points = breather.get_phi_solution()

# Using the scipy lagrange function
lagrange_polynomial = lagrange(x_points, phi_points)
print("Using the scipy function we get \n",lagrange_polynomial)

# Using the breather solution to find the true function
x_true = np.linspace(a,b,100)
true_breather = Breather_Solution(x_true)
true_phi_points = true_breather.get_phi_solution()

# Plotting true function and lagrange polynomial on the same plot
true_breather.plot_phi_solution('-','true function')
plt.plot(x_true, lagrange_polynomial(x_true), '--', label="Lagrange polynomial")
breather.plot_phi_solution('o', 'points given')
plt.title('Testing Lagrange Polynomial with 12 Gauss Lobato Nodes ')
plt.legend()
plt.grid();
```

I also created an `energy_function` function which held the integral given above. This function calls my `get_solution` function which is the $\phi(x,t)$ bit of the energy function given above. Once I had the $\phi$ integration points through using the energy function, I used `scipy.integrate.simpson` to estimate the integral, which is printed below.

```python
def get_solution(x, t = 0, omega = 0.5):
        phi = 4 * np.arctan(np.sqrt(1 - omega**2) * np.cos(omega * t) / (omega * np.cosh(np.sqrt(1 - omega**2) * x)))
        return phi

# Implementing the energy function formula given above
def energy_function(x,t):
    integral = 1 - np.cos(get_solution(x,t))
    return integral
    
# Finding the x points
a = -15
b = 15
N = 11
t = 0
x_integration = np.linspace(a , b,  N)

# Finding the energy function
phi_integration = energy_function(x_integration,t)

# Integrating the energy function
integral_scipy = simpson(phi_integration,x_integration)
print('The integral with', N , 'points =', integral_scipy)
```

The code to generate the pseudospectral derivative matrices for the first and second derivatives of any function $f(x)$ for a grid containing $N$ points *at the Gauss Lobato points* in $x$ is given below for the standard interval $[-1,1]$, and verified on some example functions.

```python
# Code provided for the derivative matrices at Gauss Lobato nodes for the interval [-1,1]

def get_pseudospectral_first_derivative_matrix(N) :
    
    indices = np.arange(N+1)
    u_i = np.cos(np.pi * indices / N)
    c_i = np.ones_like(indices)
    c_i[0] = 2.0
    c_i[N] = 2.0

    D_matrix = np.zeros([N+1,N+1])

    for idx_i in indices :
        for idx_j in indices :
            if(idx_i == 0 and idx_j == 0) : 
                D_matrix[idx_i,idx_j] = (2.0 * N * N + 1.0)/6.0
                
            elif (idx_i == N and idx_j == N) :
                D_matrix[idx_i,idx_j] = -(2.0 * N * N + 1.0)/6.0
                
            elif (idx_i == idx_j) :
                D_matrix[idx_i,idx_j] = - u_i[idx_i] / 2.0 / (1.0 - u_i[idx_i] * u_i[idx_i])
                
            else :
                D_matrix[idx_i,idx_j] = (c_i[idx_i] / c_i[idx_j] * (-1)**(idx_i+idx_j) 
                                         / (u_i[idx_i] - u_i[idx_j]))
    return D_matrix

def get_pseudospectral_second_derivative_matrix(N) :
    
    indices = np.arange(N+1)
    u_i = np.cos(np.pi * indices / N)
    c_i = np.ones_like(indices)
    c_i[0] = 2.0
    c_i[N] = 2.0

    D2_matrix = np.zeros([N+1,N+1])

    for idx_i in indices :
        for idx_j in indices :
            if((idx_i == idx_j) and (idx_i == 0 or idx_i == N)) : 
                D2_matrix[idx_i,idx_j] = (N ** 4.0 - 1.0) / 15.0
            
            elif ((idx_i == idx_j) and (idx_i != 0) and (idx_i != N)) :
                D2_matrix[idx_i,idx_j] = (- ((N*N - 1.0) * (1.0 - u_i[idx_i] * u_i[idx_i]) + 3.0) 
                                             / (3.0 * (1.0 - u_i[idx_i] * u_i[idx_i]) ** 2.0))
                 
            elif (idx_i == 0 and idx_j > 0) :
                D2_matrix[idx_i,idx_j] = ((2.0 / 3.0) * ((-1) ** idx_j) / c_i[idx_j] 
                                          * ((2.0 * N * N + 1.0) * (1.0 - u_i[idx_j]) - 6.0)
                                          / (1.0 - u_i[idx_j]) / (1.0 - u_i[idx_j]))
                
            elif (idx_i == N and idx_j < N) :
                D2_matrix[idx_i,idx_j] = ((2.0 / 3.0) * ((-1) ** (N+idx_j)) / c_i[idx_j] 
                                          * ((2.0 * N * N + 1.0) * (1.0 + u_i[idx_j]) - 6.0)
                                          / (1.0 + u_i[idx_j]) / (1.0 + u_i[idx_j]))
                
            else :
                D2_matrix[idx_i,idx_j] = (((-1) ** (idx_i+idx_j)) / c_i[idx_j] 
                                            * (u_i[idx_i]*u_i[idx_i] + u_i[idx_i]*u_i[idx_j] - 2.0)
                                            / (1.0 - u_i[idx_i]*u_i[idx_i]) / (u_i[idx_i] - u_i[idx_j])
                                            / (u_i[idx_i] - u_i[idx_j]))
                
    return D2_matrix
```

- **First and Second Derivative Matrices**

Below I have implemented my stencils into a 1st and 2nd FD derivative matrix. My `get_FD_derivative_matrix` function first finds the step size (h) and then `one_over_2h` to make the calculations more easily readable. I then input my stencil for the 3 different points: if $i == j$ the value is $0$, if $i == j + 1$ the value is $-1 / 2h$, if $i == j - 1$ the value is $1 / 2h$. I then fitted the ends with the one sided derivatives which I found at https://web.media.mit.edu/~crtaylor/calculator.html. I then followed the exact same process for the `get_FD_second_derivative_matrix` function but just with the other stencil found above.

```python
def get_FD_derivative_matrix(N) :
    
    h = 2.0 / N #finding the equal step sizes over an interval of 2 
    one_over_2h = 1.0 / (2*h) # to make the code easier to understand
    D_matrix = np.zeros([N+1,N+1]) # create the matrix

    indices = np.arange(N+1)
    # plugging the values of our first derivative stencil in the question above 
    for idx_i in indices :
        for idx_j in indices:
            if (idx_i == (idx_j)) :
                D_matrix[idx_i,idx_j] = 0.0 # f2 the middle 
            
            elif (idx_i == (idx_j+1)) :
                D_matrix[idx_i,idx_j] = -1.0 * one_over_2h # f1 
                
            elif (idx_i == (idx_j-1)) :
                D_matrix[idx_i,idx_j] = 1.0 * one_over_2h # f3
            
            else :
                D_matrix[idx_i,idx_j] = 0.0
                
    # Fix the ends with one sided derivatives
    D_matrix[0,0] = -3.0 * one_over_2h
    D_matrix[0,1] = 4.0 * one_over_2h
    D_matrix[0,2] = -one_over_2h
    D_matrix[N,N] = 3.0 * one_over_2h
    D_matrix[N,N-1] = -4.0 * one_over_2h
    D_matrix[N,N-2] = one_over_2h
    
    return D_matrix

def get_FD_second_derivative_matrix(N) :
    
    h = 2.0 / N 
    one_over_h2 = 1.0 / (h*h) # just for readability
    D2_matrix = np.zeros([N+1,N+1])

    indices = np.arange(N+1)

    # plugging the values of our second derivative stencil in the question above 
    for idx_i in indices :
        for idx_j in indices :
            if(idx_i == idx_j) :
                D2_matrix[idx_i,idx_j] = -2.0 * one_over_h2 # f2 the middle
                
            elif (idx_i == (idx_j+1)) :
                D2_matrix[idx_i,idx_j] = 1.0 * one_over_h2 # f1
                
            elif (idx_i == (idx_j-1)) :
                D2_matrix[idx_i,idx_j] = 1.0 * one_over_h2 #f3
                
            else :
                D2_matrix[idx_i,idx_j] = 0.0
                
    # Fix the ends with one sided derivatives
    D2_matrix[0,0] = 1.0 * one_over_h2
    D2_matrix[0,1] = -2.0 * one_over_h2
    D2_matrix[0,2] = 1.0 * one_over_h2
    D2_matrix[N,N] = 1.0 * one_over_h2
    D2_matrix[N,N-1] = -2.0 * one_over_h2
    D2_matrix[N,N-2] = 1.0 * one_over_h2
    
    return D2_matrix
```

**Q4a**. Explain what type of PDE Eq. (1) is and discuss how this impacts on well-posedness and the way in which we can solve the equation numerically.

- **What type of PDE is Eq. (1)?**

Equation (1):

$$ \frac{\partial\varphi^2}{\partial t^2} -\frac{\partial\varphi^2}{\partial x^2}+\sin \varphi =0, \quad$$

 is a non-linear hyperbolic PDE for a function $\varphi$ which is dependent on $x$ and time (t). Using the general PDE equation:

$$A \frac{\partial^2 u}{\partial x^2} + 2B \frac{\partial^2 u}{\partial x \partial y} + C \frac{\partial^2 u}{\partial y^2} + D \frac{\partial u}{\partial x} + E \frac{\partial u}{\partial y} + F = 0$$

We can see that for equation (1): $a = -1$, $c = 1$ and $B = 0$. Using the discriminant: $B^2 - 4ac$ we get the discriminant equal to 4 resulting in the PDE being hyperbolic as it is greater than 0.

- **How does this impact Well-posedness?**

If we make the Sine Gordon equation purely first order, we obtain the following system (I have replaced $\varphi$ with u):

$\frac{\partial u}{\partial t} = v,$ $\frac{\partial v}{\partial t} = \frac{\partial\psi}{\partial x}$, $\frac{\partial\psi}{\partial t} = \frac{\partial v}{\partial x} - sin(u)$

$\frac{\partial}{\partial t}\begin{bmatrix}
    u \\
    v \\
    \psi \\
\end{bmatrix} = 
\begin{bmatrix}
    0 & 0 & 0 \\
    0 & 0 & 1 \\
    0 & 1 & 0 \\
\end{bmatrix}
\frac{\partial}{\partial x}\begin{bmatrix}
    u \\
    v \\
    \psi \\
\end{bmatrix} + 
\begin{bmatrix}
    v \\
    0 \\
    0 \\
\end{bmatrix}
$ $ - sin(u)$

The characteristic matrix here has eigenvalues: 0, 1, -1. As the eigenvalues are all distinct the system is strictly hyperbolic and therefore well-posed.

- **How do we solve the equation numerically?**

To solve this equation numerically I have defined the time derivative of $\varphi$ as (note I have swapped $\varphi$ with u):

$$v = \frac{\partial u}{\partial t}$$

This allows us to solve the coupled system:

$$\frac{\partial v}{\partial t} = \frac{\partial u^2}{\partial x^2} - sin(u),$$          

$$\frac{\partial u}{\partial t} = v$$

I then represented this in matrix form, where $\frac{\partial v}{\partial t}$ equals the dot product of Matrix $D^2$ and Vector $u$ $- sin(u)$ and $\frac{\partial u}{\partial t}$ equals the Vector $v$.

 - y0 is the initial conditions which will be inputted when instantiating the class
 - t_eval specifies which time points to store the results which here will be the t solution


**Integrate Using the Midpoint Method**

 The midpoint method is another way to solve PDEs which is modified from the Euler method. This method is an improvement on the Euler method as it achieves stability by using the gradient at a half time step and using that to compute the full step gradient. 

I start by working out the y value at the midpoint using this formula:

$$y_{k+\frac{1}{2}} = y_k + \frac{1}{2} \cdot h \cdot f(y_k, t_k)$$
where$$f = \frac{dy}{dt}$$

Which is put into my code here: 
~~~
y_t_plus_half = y_solution[:, itime-1] + 0.5 * dydt_at_t * delta_t
~~~

Then after calculating $\frac{dy}{dt}$ at t plus half, I used this formula to find the y value at the full step:

$$y_{k+1} = y_k + h \cdot f(y_{k+\frac{1}{2}}, t_{k+\frac{1}{2}})$$

Which is used in my code here:
~~~
y_solution[:, itime] = (y_solution[:, itime-1] + dydt_at_t_plus_half * delta_t) 
~~~

**Integrate Using Trapezoid Rule**

This function implementes the formula: 

$$y_{k+1} = y_k + \frac{h}{2} \left( f(y_{k+1}, t_{k+1}) + f(y_k, t_k) \right)
$$

Where $f(y_k, t_k)$ is the derivative at the current time step and $f(y_{k+1}, t_{k+1})$ is the derivative at the next time step. Therefore the trapezoidal rule takes the average of these two values and adds onto the previous value of $y_k$.

```python
class Sine_Gordon_PDE :
    
    """
    Represents the Sine Gordon Equation PDE and its solution
    
    Attributes: CFL, x and y0 (initial conditions),
                N_grid (the size of the grid)
    
    Methods: Integration using solve ivp and midpoint method, calculate dydt
    
    """
    
    # constructor function
    def __init__(self, x,y0, max_time, CFL=1.0, method="finite_difference"):
        self.CFL = CFL
        self.max_time = max_time
        self.x = x
        self.y0 = y0
        self.N_grid = np.size(x)            
        self._solution_t = []
        self._solution_y = []
        self.method = method
        
        self.initial_value = self.y0
        
        if self.method == "finite_difference" :
            self.D2_matrix = get_FD_second_derivative_matrix(self.N_grid-1)
        elif self.method == "pseudospectral" :
            self.D2_matrix = get_pseudospectral_second_derivative_matrix(self.N_grid-1)
        else :
            assert False, "Requested derivative method not implemented"
            
        # Rescale for the derivative matrix for the interval we are looking at
        b_minus_a = self.x[self.N_grid-1] - self.x[0]
        self.D2_matrix *= 4.0 / b_minus_a / b_minus_a
    
    def calculate_dydt(self, t, current_state) :
        
        # Just for readability
        [u,v] = np.array_split(current_state, 2)
        dydt = np.zeros_like(current_state)
        dudt, dvdt = np.array_split(dydt, 2)
        
        # Work out the time derivatives
        dudt[:] =  v
        dvdt[:] =  np.dot(self.D2_matrix, u) - np.sin(u)
           
        # Zero the derivatives at the end for stability 
        dydt[0] = 0.0
        dydt[1] = 0.0
        dydt[self.N_grid-1] = 0.0
        dydt[self.N_grid-2] = 0.0
        dydt[self.N_grid] = 0.0
        dydt[self.N_grid+1] = 0.0
        dydt[self.N_grid*2-1] = 0.0
        dydt[self.N_grid*2-2] = 0.0
        
        return dydt      

    def integrate_using_solve_ivp(self) :
        
        t_solution = np.linspace(0.0, self.max_time, 100)
        solution = solve_ivp(self.calculate_dydt, [0, self.max_time],
                        self.initial_value, t_eval=t_solution)
        
        self._solution_t = solution.t
        self._solution_y = solution.y
        
        return solution.t, solution.y

    def integrate_using_midpoint(self) :
        
        assert self.method == "finite_difference", 'need to use FDs'
        dx = self.x[1] - self.x[0] # assume constant using FDs
        
        # Using the CFL condition in practise just to be sure it is stable
        delta_t = self.CFL * dx
        num_timesteps = int(self.max_time / delta_t) + 1
        t_solution = np.linspace(0.0, self.max_time, num_timesteps)
        y_solution = np.zeros([2*self.N_grid, num_timesteps])
        
        for itime, time in enumerate(t_solution) :
            if itime == 0 :
                y_solution[:, itime] = self.initial_value
                old_time = time
            else :
                dydt_at_t = self.calculate_dydt(old_time, y_solution[:, itime-1])
                
                # Take half step
                y_t_plus_half = y_solution[:, itime-1] + 0.5 * dydt_at_t * delta_t
                t_plus_half = old_time + 0.5 * delta_t
                dydt_at_t_plus_half = self.calculate_dydt(t_plus_half, y_t_plus_half)
                
                # Take full step
                y_solution[:, itime] = (y_solution[:, itime-1] + dydt_at_t_plus_half * delta_t) 
                old_time = time 

        self._solution_t = t_solution
        self._solution_y = y_solution
        
        return t_solution, y_solution
    
    def integrate_using_trapezoidal_method(self, a_num_steps=1000) :
        
        assert self.method == "finite_difference" or "pseudospectral"
        t_solution = np.linspace(0.0, self.max_time, a_num_steps)
        y_solution = np.zeros([2*self.N_grid, a_num_steps])
        delta_t = self.max_time / (a_num_steps - 1)
        
        for itime, time in enumerate(t_solution) :
            if itime == 0 :
                y_solution[:,itime] = self.initial_value
                old_time = time
            else :
                y_of_t_old   = y_solution[:,itime-1]
                dydt_at_t = self.calculate_dydt(old_time, y_solution[:, itime-1])
                # Remember we have to copy if we don't want to actually
                # amend the elements of y_of_t_old
                y_of_t_guess = np.copy(y_of_t_old)
                
                # Implement the iterative scheme described in the lecture
                error = 10.0
                error_threshold = 1e-1
                while (error > error_threshold) :
                    
                    # Get the value of dydt at the guessed value
                    dydt_at_t_next = self.calculate_dydt(time, y_of_t_guess)
                    
                    # Get the new guess using the Euler update rule
                    y_of_t_new = y_of_t_old + delta_t/2 * (dydt_at_t_next + dydt_at_t)
                    
                    # If the new guess is the same as the old one we have converged
                    error = np.linalg.norm(y_of_t_guess - y_of_t_new)
                    y_of_t_guess = y_of_t_new
                    
                    # Check it is not getting worse!
                    assert error < 20.0, 'Non linear iterations not converging!'
                
                # Now (assuming it has converged) assign the 
                # value we found to the solution and continue the time iteration
                # Enforce the boundary conditions
                y_of_t_guess[0] = 0.0
                y_of_t_guess[self.N_grid-1] = 0.0
                y_solution[:,itime] = y_of_t_guess
                
        self._solution_t = t_solution
        self._solution_y = y_solution
        
        return t_solution, y_solution
```

- **Solve Using Finite Differences and Midpoint Method**

PDE's can often be unstable which results in a 'CFL condition' on the time step. What this condition is telling us is that if we take a step in time, we must take a similiar step in space. For hyperbolic equations the steps should be linearly related:

$$Δt = λΔx$$

I played around with $\lambda$ and found that when it was 1 the solution wasn't very stable but as I reduced it down to 0.5 it gave much nicer and more stable plots. As you can see by the plot the solution remains stationary.

```python
num_points = 80
a=-6
b=6
# Find evenly spaced x points as we are using FD 
x_values = np.linspace(a , b,  num_points)
breather = Breather_Solution(x_values)

# As shown above v = dphi/dt
v0 = breather.get_dphidt_solution()
phi0 = breather.get_phi_solution()

# Join together v0 and phi0 
y0 = np.concatenate([phi0,v0])

# Setting initial conditions
max_time = 6
CFL = 0.5

# Calling my class and using the midpoint method
my_gordon_equation = Sine_Gordon_PDE(x_values, y0, max_time, CFL, method = 'finite_difference')
t_solution, y_solution = my_gordon_equation.integrate_using_midpoint()
phi_solution = y_solution[0:num_points,:]

for i, t_i in enumerate(t_solution) :
    if (i > 0) and (i % 5 == 0) and (t_i > 0.0):
        labelt = "t="+str(round(t_i,2))
        f_t = phi_solution[:,i]
        plt.plot(x_values, f_t, label=labelt, alpha=(1-t_i/max_time))

plt.xlabel('x')
plt.ylabel('value over time of phi')
plt.title('Solving Sine Gordon Equation Using FD and Midpoint')
plt.grid()
```

- **Solve Using Pseudospectral Derivative and Solve ivp**

*Advantages and Disadvantages of using the pseudospectral derivatives:*

In general pseudospectral methods are more accurate but harder to calculate. For example, the error has an exponential convergence rate which is faster than the polynomial rate. But we have to find the gauss lobato nodes instead of evenly spaced points and all the values of the function are used in the stencil and not just the neighbouring ones which makes the matrix much harder to calculate.

*What is the time integration method usually used by solve ivp and how does it avoid potential instabilities?*

Solve-ivp uses the Runge-Kutta time integration method, inparticular RK45. Runge-Kutta methods acheives stability by using even more intermediate steps than the midpoint method to calculate the full time step. Solve-ivp takes a 4th order and then a 5th order step and calculates the difference (step error). If this step error is greater than rtol (some threshold) then it reduces the step sizes.


```python
num_points = 28
a=-6
b=6

# Finding the gauss lobato nodes as we are using pseudospectral method
u_i = np.cos(np.pi * np.arange(num_points+1) / num_points)
x_values = (b - a)/2.0 * u_i + (a + b)/2.0
breather = Breather_Solution(x_values)

# As shown above v = dphi/dt
phi0 = breather.get_phi_solution()
v0 = breather.get_dphidt_solution()

# Join together v0 and phi0 
y0 = np.concatenate([phi0,v0])

# Setting the initial conditions
max_time = 5
CFL = 1.0

# Calling my class and using the solve ivp method
my_gordon_equation = Sine_Gordon_PDE(x_values, y0, max_time, CFL, method = 'pseudospectral')
t_solution, y_solution = my_gordon_equation.integrate_using_solve_ivp()

u_solution = y_solution[0:num_points+1,:]

for i, t_i in enumerate(t_solution) :
    if (i > 0) and (i % 5 == 0) and (t_i > 0.0):
        labelt = "t="+str(round(t_i,2))
        f_t = u_solution[:,i]
        
        # Just to make the plot nice use the points to reconstruct the fitting function
        lagrange_polynomial = lagrange(x_values, f_t)
        x_lagrange = np.linspace(a, b, 100)
        plt.plot(x_lagrange, lagrange_polynomial(x_lagrange), '-', 
                 label=labelt, alpha=(1-t_i/max_time))

plt.xlabel('x')
plt.ylabel('value over time of phi')
plt.title('Solving Sine Gordon Equation Using Pseudospectral and Solve ivp')
plt.grid()
```
