The goal of the `standardise` function is to make sure different features of objects are on almost on the same scale so that each feature is equally important and make it easier to process by most ML algorithms. The result of standardisation is that the features will be rescaled to ensure the mean and the standard deviation to be $0$ and $1$, respectively.

```python
def standardise(data_matrix):
    row_of_means = np.mean(data_matrix, axis=0)
    standardised_matrix = data_matrix - row_of_means
    row_of_stds = np.std(standardised_matrix, axis=0)
    return (standardised_matrix / row_of_stds), row_of_means, row_of_stds

def de_standardise(standardised_matrix, row_of_means, row_of_stds):
    matrix = np.copy(standardised_matrix * row_of_stds)
    return matrix + row_of_means
```

# **Part 2: Modelling**

The `linear_regression_data` function coverts the data inputs variables above into a data matrix as so:
$$
\mathbf{\Phi\left(X\right)} = 
\begin{pmatrix}
1 & x^{(1)}_1 & x^{(1)}_2 & \ldots & x^{(1)}_d \\
1 & x^{(2)}_1 & x^{(2)}_2 & \ldots & x^{(2)}_d \\
\vdots & \vdots & \vdots & \ddots & \vdots & \\
1 & x^{(s)}_1 & x^{(s)}_2 & \ldots & x^{(s)}_d \\
\end{pmatrix}
$$

```python
def linear_regression_data(data_inputs):
    first_column = np.ones((len(data_inputs), 1))
    X_matrix = np.c_[first_column, data_inputs]
    return X_matrix

def model_function(data_matrix, weights):
    return data_matrix @ weights
```

The `logistic_function` function returns the sigmoid function $\sigma(x)$:

\begin{equation*} 
\sigma(x): = \frac{1}{1+\mathrm{e}^{-x}} \, , 
\end{equation*}

where $x$ is the data inputs

```python
def logistic_function(inputs):
    return 1 / (1 + np.exp(-inputs))

def binary_prediction_function(data_matrix, weights):
    # the binary classification can be obtained by applying the logistic function to the model function
    probability = logistic_function(model_function(data_matrix, weights))
    return probability > 1/2
```

The `classification_accuracy` function takes the predicted labels from the `binary_prediction_function` and true labels to evaluate what percentage is the same.

```python
def classification_accuracy(estimated_labels, true_labels):
    equal_labels = estimated_labels == true_labels # this returns an array made of true and false 
    return np.mean(equal_labels) # the average is how many true we have
```

The `gradient_descent` function minimises the function so that it returns the optimal weights and objective values for each iteration.

```python
def gradient_descent(objective,gradient,initial_weights,step_size=1,no_of_iterations=100):
    
    objective_values = []
    weights = np.copy(initial_weights)
    objective_values.append(objective(weights))
    
    for counter in range(no_of_iterations):
        weights -= step_size * gradient(weights)
        objective_values.append(objective(weights))
        #if (counter + 1) % print_output == 0:
         #   print("Iteration {k}/{m}, objective = {o}.".format(k=counter+1,m=no_of_iterations, 
          #                                                     o=objective_values[counter]))
            
    #print("Iteration completed after {k}/{m}, objective = {o}.".format(k=counter+ 1,
     #                                                                  m=no_of_iterations, 
      #                                                                 o=objective_values[counter]))
    return weights, objective_values
```

The `binary_logistic_regression_cost_function` implements the cost function for binary logisitic regression, which is:

$$
\mathrm{L}\left(\mathbf{w}\right) = \frac{1}{s} \left(
\sum\limits_{i=1}^s \log\left[1+\exp\left(f\left(\mathbf{x}^{(i)},\mathbf{w}\right)\right)\right] - y_i\cdot f\left(\mathbf{x}^{(i)},\mathbf{w}\right)
\right),
$$

```python
def binary_logistic_regression_cost_function(data_matrix, data_labels,weights):
    regression_outputs = model_function(data_matrix, weights)
    return np.mean(np.log(1 + np.exp(regression_outputs)) -data_labels * regression_outputs)
```

The `binary_logistic_regression_gradient` implements the gradient for binary logisitic regression, which is:

$$
\nabla \mathrm{L}\left(\mathbf{w}\right) = \frac{1}{s} \left(
\sum\limits_{i=1}^s \phi\left(\mathbf{x}^{(i)}\right)\cdot\sigma
\left(\left\langle \phi\left(\mathbf{x}^{(i)}\right),\mathbf{w} \right\rangle \right) - y_i\cdot \phi\left(\mathbf{x}^{(i)}\right)
\right),
$$

```python
def binary_logistic_regression_gradient(data_matrix, data_labels, weights):
    return data_matrix.T @ (logistic_function(model_function(data_matrix, weights)) - data_labels) / len(data_matrix)
```

```python
# Get the data matrix from the bank input
bank_data_matrix = linear_regression_data(bank_input)

# Define  cost function
bank_cost_function = lambda weights: binary_logistic_regression_cost_function(bank_data_matrix,
                                                                                 bank_labels, weights)
# Define the gradient
bank_gradient_function = lambda weights: binary_logistic_regression_gradient(bank_data_matrix,
                                                                                bank_labels, weights)

# Initial weights of w0
bank_initial_weights = np.zeros((len(bank_data_matrix.T), 1))

# Setting step size 
bank_step_size = 3.9 * len(bank_data_matrix) / (np.linalg.norm(bank_data_matrix))**2

# Gradient descent
bank_optimal_weights, bank_objective_values =gradient_descent(bank_cost_function,
                                                                     bank_gradient_function,
                                                                     bank_initial_weights,
                                                                     bank_step_size,200)
# Predicted labels (churn)
predicted_labels= binary_prediction_function(bank_data_matrix,bank_optimal_weights)

# Comparing predicted v real labels
bank_accuracy_rate = classification_accuracy(predicted_labels,bank_labels)
```

```python
import random as rd
def boostrap_regression(standardised_data_input, standardised_data_output,fraction,M):
    # Find the number of samples to extract
    data_size = len(standardised_data_output)
    samples_size = int(data_size*fraction)

    accuracy_list = []
    w_list=[]
    # For each of M extract:
    for j in range(M):
        sample_input_list=[]
        sample_output_list=[]
        for i in range(samples_size):
            # Extract N samples of random numbers which are the id of the arrays that store the data
            id_random = rd.randint(0,samples_size-1)
            sample_input_list.append(standardised_data_input[id_random])
            sample_output_list.append(standardised_data_output[id_random])
        
        #convert the list to arrays
        sample_input=np.array(sample_input_list)
        sample_output=np.array(sample_output_list)

        # Apply the regression as above
        bank_data_matrix = linear_regression_data(sample_input)
        bank_cost_function = lambda weights: binary_logistic_regression_cost_function(bank_data_matrix,
                                                                                      sample_output, weights)                                                                               
        bank_gradient_function = lambda weights: binary_logistic_regression_gradient(bank_data_matrix,
                                                                                sample_output, weights)
        bank_initial_weights = np.zeros((len(bank_data_matrix.T), 1))
        bank_step_size = 3.9 * len(bank_data_matrix) / (np.linalg.norm(bank_data_matrix))**2
        bank_optimal_weights, bank_objective_values =gradient_descent(bank_cost_function,bank_gradient_function,bank_initial_weights,bank_step_size,200)
        predicted_labels= binary_prediction_function(bank_data_matrix,bank_optimal_weights)
        # calculate the accuracy of the model woth each sample and appened to a list
        bank_accuracy_rate = classification_accuracy(predicted_labels,sample_output_list)
        accuracy_list.append(bank_accuracy_rate)
        # apply the fitted values of weights for the N samples in the list
        w_list.append(bank_optimal_weights[1:])
    return w_list, accuracy_list
```
